```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_clean = df.fillna("Unknown")
```

```python
df_clean
```

```python
df_clean[df_clean["date_added"]=="Unknown"]
```

```python
df_clean.loc[df_clean["date_added"] == "Unknown", "date_added"] = "January 1, " + df_clean.loc[df_clean["date_added"] == "Unknown", "release_year"].astype(str)
```

```python
df_clean['date_added'] = df_clean['date_added'].str.strip()

df_clean["date_added"] = pd.to_datetime(df_clean["date_added"], format='%B %d, %Y')
```

```python
df_clean
```

```python
df_clean = df_clean.drop(columns=['description'], axis=1)
```

```python
df_clean["release_year"]=df_clean["release_year"].astype(int)
```
